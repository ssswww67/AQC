```python
import math
import random

# Number to factor
N = 91
flag = 0
# Step 1: Pick a random 'a' (1 < a < N)
while flag!=1:
    a = random.randint(1, N)
    while math.gcd(a, N) != 1:
        a = random.randint(2, N-1)

    print("Chosen a:", a)

    # Step 2: Find order 'r' (smallest number such that a^r ≡ 1 mod N)
    def order(a, N):
        r = 1
        while pow(a, r, N) != 1:
            r += 1
        return r

    r = order(a, N)
    print("Order r:", r)

    # Step 3: If r is even, compute possible factors
    if r % 2 == 0:
        factor1 = math.gcd(pow(a, r//2) - 1, N)
        factor2 = math.gcd(pow(a, r//2) + 1, N)
        if factor1 not in [1, N] and factor2 not in [1, N]:
            print("Factors of", N, "are:", factor1, "and", factor2)
            flag=1;

        else:
            print("Failed, try again with different a")
            print("                                  ")
            print("----------------------------------")
    else:
        print("Order r is odd, try again")
        print("                                  ")
        print("----------------------------------")
```
